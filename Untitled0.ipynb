#IMPORT THƯ VIỆN

```python
import numpy as np
import random
import time
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

#KHAI BÁO ACTION SPACE & HÀM TIỆN ÍCH

```python
ACTION_SPACE = ['UP', 'DOWN', 'LEFT', 'RIGHT']

ACTION_TO_DELTA = {
    0: (-1, 0),  # UP
    1: (1, 0),   # DOWN
    2: (0, -1),  # LEFT
    3: (0, 1)    # RIGHT
}

def print_map(map_data):
    for row in map_data:
        print(' '.join(row))
```

#KHAI BÁO MAP

```python
# Map 4x4
MAP_4x4 = [
    list("SFFF"),
    list("FHFH"),
    list("FFFH"),
    list("HFFG")
]

# Map 10x10 (khó)
MAP_10x10 = [
    list("SFFFFFFFFF"),
    list("FHFHHHFHFF"),
    list("FFFHFFFFHF"),
    list("HFFHFHFHFF"),
    list("FFFHFFFHFF"),
    list("HFHFHFFHFH"),
    list("FFFHFFFFHF"),
    list("FFHFHFHFFF"),
    list("FHFHFFFHFF"),
    list("FFFFFHFFFG")
]

# Map 12x12 (rất khó)
MAP_12x12 = [
    list("SFFFFFFFFFFF"),
    list("FHFHHHFHFFFF"),
    list("FFFHFFFFHFHF"),
    list("HFFHFHFHFFHF"),
    list("FFFHFFFHFFFF"),
    list("HFHFHFFHFHFH"),
    list("FFFHFFFFHFHF"),
    list("FFHFHFHFFFFF"),
    list("FHFHFFFHFFHF"),
    list("FFFFFHFFHFHF"),
    list("FHFHFFFFHFHF"),
    list("FFFFFFFHFFFG")
]
```

##Print map

```python
print("=== MAP 4x4 ===")
print_map(MAP_4x4)

print("\n=== MAP 10x10 ===")
print_map(MAP_10x10)

print("\n=== MAP 12x12 ===")
print_map(MAP_12x12)
```

#ĐỊNH NGHĨA ENVIRONMENT

```python
class GridWorld:
    def __init__(self, map_data):
        self.map = map_data
        self.rows = len(map_data)
        self.cols = len(map_data[0])
        self.reset()

    def reset(self):
        self.row, self.col = 0, 0
        return self.row * self.cols + self.col

    def step(self, action):
        dr, dc = ACTION_TO_DELTA[action]
        new_r = max(0, min(self.rows - 1, self.row + dr))
        new_c = max(0, min(self.cols - 1, self.col + dc))

        self.row, self.col = new_r, new_c
        cell = self.map[new_r][new_c]

        reward = -1
        done = False

        if cell == 'H':
            reward = -100
            done = True
        elif cell == 'G':
            reward = 100
            done = True

        next_state = self.row * self.cols + self.col
        return next_state, reward, done
```

#Q-LEARNING AGENT

```python
class QLearningAgent:
    def __init__(self, state_size, action_size):
        self.q_table = np.zeros((state_size, action_size))
        self.lr = 0.1
        self.gamma = 0.9
        self.epsilon = 1.0
        self.epsilon_decay = 0.995
        self.epsilon_min = 0.01

    def choose_action(self, state):
        if random.random() < self.epsilon:
            return random.randint(0, 3)
        return np.argmax(self.q_table[state])

    def update(self, state, action, reward, next_state):
        best_next = np.max(self.q_table[next_state])
        self.q_table[state, action] += self.lr * (
            reward + self.gamma * best_next - self.q_table[state, action]
        )
```

#HUẤN LUYỆN TỪNG MAP

```python
def train_and_test(map_data, episodes=2000, max_steps=300):
    env = GridWorld(map_data)
    agent = QLearningAgent(env.rows * env.cols, 4)

    # TRAIN
    for ep in range(episodes):
        state = env.reset()
        for _ in range(max_steps):
            action = agent.choose_action(state)
            next_state, reward, done = env.step(action)
            agent.update(state, action, reward, next_state)
            state = next_state
            if done:
                break
        agent.epsilon = max(agent.epsilon * agent.epsilon_decay, agent.epsilon_min)

    # TEST + SHOW PATH
    print("\n=== TEST AGENT ===")
    state = env.reset()
    path = []

    for step in range(max_steps):
        action = np.argmax(agent.q_table[state])
        next_state, reward, done = env.step(action)

        path.append((env.row, env.col, ACTION_SPACE[action]))
        state = next_state
        if done:
            break

    for i, p in enumerate(path):
        print(f"Bước {i+1:02d} | Vị trí: ({p[0]}, {p[1]}) | Action: {p[2]}")
```

#DEMO ĐƯỜNG ĐI

```python
print("===== CHẠY MAP 4x4 =====")
train_and_test(MAP_4x4)

print("\n\n===== CHẠY MAP 10x10 =====")
train_and_test(MAP_10x10, episodes=4000)

print("\n\n===== CHẠY MAP 12x12 =====")
train_and_test(MAP_12x12, episodes=6000)
```

#Hàm Train có ghi lại biểu đồ

```python
def train_with_history(map_data, episodes=3000, max_steps=300):
    env = GridWorld(map_data)
    agent = QLearningAgent(env.rows * env.cols, 4)

    rewards_history = []
    steps_history = []

    for ep in range(episodes):
        state = env.reset()
        total_reward = 0

        for step in range(max_steps):
            action = agent.choose_action(state)
            next_state, reward, done = env.step(action)

            agent.update(state, action, reward, next_state)

            state = next_state
            total_reward += reward

            if done:
                break

        rewards_history.append(total_reward)
        steps_history.append(step + 1)

        agent.epsilon = max(agent.epsilon * agent.epsilon_decay, agent.epsilon_min)

    return agent, rewards_history, steps_history
```

#TRAIN RIÊNG TỪNG MAP + LƯU LỊCH SỬ

```python
print("Training MAP 12x12...")
agent_12x12, rewards_12x12, steps_12x12 = train_with_history(MAP_12x12, episodes=6000)
```

#BIỂU ĐỒ 3: TRUNG BÌNH

```python
def moving_average(data, window=100):
    return np.convolve(data, np.ones(window)/window, mode='valid')

plt.figure()
plt.plot(moving_average(rewards_12x12), label="12x12")

plt.xlabel("Episode")
plt.ylabel("Reward trung bình")
plt.title("Reward trung bình (Moving Average)")
plt.legend()
plt.grid(True)
plt.show()
```
